```python
#  Imports
import nltk
from nltk.corpus import gutenberg, stopwords
from nltk import word_tokenize
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# Download necessary NLTK data
nltk.download('gutenberg')
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
#  Load two Gutenberg texts
text1 = gutenberg.raw('austen-emma.txt')
text2 = gutenberg.raw('bible-kjv.txt')
```

```python
#  Tokenize both texts
tokens1 = word_tokenize(text1)
tokens2 = word_tokenize(text2)
```

```python
#  Remove stopwords and punctuation
stop_words = set(stopwords.words('english'))
clean1 = [w.lower() for w in tokens1 if w.isalpha() and w.lower() not in stop_words]
clean2 = [w.lower() for w in tokens2 if w.isalpha() and w.lower() not in stop_words]
```

```python
#  Generate word clouds
wc1 = WordCloud(width=800, height=400).generate(' '.join(clean1))
wc2 = WordCloud(width=800, height=400).generate(' '.join(clean2))
```

```python
#  Display word clouds
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(wc1); plt.axis('off'); plt.title('Text 1: Emma')
plt.subplot(1,2,2)
plt.imshow(wc2); plt.axis('off'); plt.title('Text 2: Bible')
plt.show()
```

```python
#  Imports
import nltk
from nltk.corpus import brown, stopwords
from nltk.stem import WordNetLemmatizer
from nltk import FreqDist
import matplotlib.pyplot as plt
```

```python
# Download necessary NLTK data
nltk.download('wordnet')
```

```python
#  Load Fiction Category
words = brown.words(categories='fiction')
```

```python
# Remove stopwords and apply lemmatization
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
clean = [lemmatizer.lemmatize(w.lower()) for w in words if w.isalpha() and w.lower() not in stop_words]
```

```python
#  Find 15 most frequent words
fdist = FreqDist(clean)
top15 = fdist.most_common(15)
```

```python
#  Plot
words_ = [w for w,c in top15]
counts = [c for w,c in top15]

plt.figure(figsize=(8,5))
plt.bar(words_, counts)
plt.xticks(rotation=45)
plt.title('Top 15 Words in Fiction Category')
plt.show()
```

```python
#  Imports
import nltk
from nltk.corpus import gutenberg, stopwords
from nltk import word_tokenize, FreqDist, bigrams
import pandas as pd
```

```python
#  Load Gutenberg text
text = gutenberg.raw('austen-emma.txt')
```

```python
#  Tokenize & clean
tokens = word_tokenize(text)
stop_words = set(stopwords.words('english'))
clean = [w.lower() for w in tokens if w.isalpha() and w.lower() not in stop_words]
```

```python
#  Find top 20 frequent words
fdist = FreqDist(clean)
top20 = [w for w,_ in fdist.most_common(20)]
```

```python
#  Generate bigrams
bigram_list = list(bigrams(clean))
```

```python
#  Build co-occurrence matrix
matrix = pd.DataFrame(0, index=top20, columns=top20)

for w1, w2 in bigram_list:
    if w1 in top20 and w2 in top20:
        matrix.loc[w1, w2] += 1

print(matrix)
```

```python
#  Imports
import nltk
from nltk.corpus import udhr
from nltk import FreqDist, ngrams
```

```python
# Download necessary NLTK data
nltk.download('udhr')
```

```python
#  Load Hungarian text
words = udhr.words('Hungarian_Magyar-Latin1')

#  Define vowels
vowels = 'aeiou'
```

```python
#  Extract vowel trigrams
vowel_trigrams = []

for w in words:
    seq = ''.join([ch.lower() for ch in w if ch.lower() in vowels])
    if len(seq) >= 3:
        vowel_trigrams.extend([''.join(t) for t in ngrams(seq, 3)])
```

```python
#  Frequency distribution
fd = FreqDist(vowel_trigrams)

# Display sorted list
for trigram, count in fd.most_common(20):
    print(trigram, count)
```
